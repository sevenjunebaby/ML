```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib

# Load the testing dataset
test_df = pd.read_csv('/home/wissal/Desktop/ML/watermelon-machine-learning/test.csv')

# Define the same features as in the training data
features = [
    'Temperature (°C)', 'Humidity (%)', 'Soil Moisture (%)', 'Soil pH', 'Sunlight (hours)',
    'Pest/Disease Incidence', 'Fertilizer Usage (kg/ha)', 'Watermelon Weight (kg)',
    'Harvest Time', 'Rainfall (mm)', 'Wind Speed (km/h)', 'CO2 Levels (ppm)', 'Nitrogen Levels (mg/L)',
    'Phosphorus Levels (mg/L)', 'Potassium Levels (mg/L)', 'Watermelon Plant Height (cm)', 
    'Watermelon Variety_Charleston Gray', 'Watermelon Variety_Crimson Sweet',
    'Watermelon Variety_Jubilee', 'Watermelon Variety_Sugar Baby', 'Geographical Location_Coastal region',
    'Geographical Location_Desert region', 'Geographical Location_Inland region', 'Geographical Location_Mountain region',
    'Geographical Location_Plains region'
]

target = 'Watermelon Disease Presence'  # Define the target variable

# Ensure the test dataset has all the necessary columns
for col in features:
    if col not in test_df.columns:
        test_df[col] = np.nan  # or use 0 or another default value

# Separate features and target variable from the test dataset
X_test = test_df[features]
y_test = test_df[target]

# Apply the same one-hot encoding to the test dataset as was done for the training dataset
X_test = pd.get_dummies(X_test, columns=[
    'Watermelon Variety_Charleston Gray', 'Watermelon Variety_Crimson Sweet',
    'Watermelon Variety_Jubilee', 'Watermelon Variety_Sugar Baby', 
    'Geographical Location_Coastal region', 'Geographical Location_Desert region',
    'Geographical Location_Inland region', 'Geographical Location_Mountain region',
    'Geographical Location_Plains region'
])

# Load the trained Decision Tree model
dt_model = joblib.load('/home/wissal/Desktop/ML/watermelon-machine-learning/decision_tree_model.joblib')

# Ensure all one-hot encoded columns are present in the test set
for col in dt_model.feature_names_in_:
    if col not in X_test.columns:
        X_test[col] = 0  # add missing one-hot encoded columns with default value 0

# Align columns to match the model's training data
X_test = X_test[dt_model.feature_names_in_]

# Make predictions on the test data
y_pred = dt_model.predict(X_test)

# Evaluate the model's performance on the test data
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(classification_report(y_test, y_pred))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))

```
